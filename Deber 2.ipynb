**Output:**
- True
- 2

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next_= None
        
class Singly_linked_list:
   
    def __init__(self):
        self.head_node = None
        self.nodos = 0
        
#Insertar al final
    def insertfin(self,val):
        nuevoN= Node(val)
        if (not self.head_node):
            self.head_node = nuevoN
        else:
            a= self.head_node
            
            while(a.next_):
                a = a.next_
            a.next_ = nuevoN
        self.nodos = self.nodos+1
        
    #Longitud 
    def longitud(self):
        return self.nodos 
    
    def loop(self, n):
        L = self.head_node
        
        for n in range (1, n):
            L = L.next_
            
        Null = self.head_node
        while (Null.next_):
            Null = Null.next_
        Null.next_ = L
    
    def traverse(self):
        P1 = self.head_node
        P2 = self.head_node
        while P1 and P2 and P2.next_:
            P1 = P1.next_
            P2 = P2.next_.next_
            if(P1 == P2):
                print("\n Cycle [2] ", P1.val)
                return True        
        return False
    
    
    def listprint(self):
        printval = self.head_node
        
        print("Lista["+str(self.nodos)+"]", end= "-> ")
        
        while printval != None:
            print (printval.val, end =" => ")
            printval = printval.next_
        print("Cycle")
        
lista = Singly_linked_list()

lista.insertfin("Jan")
lista.insertfin("Feb")
lista.insertfin("Mar")
lista.insertfin("Dic")
lista.listprint()
lista.loop(3)
lista.traverse()
```

```python
import ctypes
class Queue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.item_count1 = 0
        self.item_count2 = n-1
        self.n = n
        self.queue = self._create_queue(self.n)        
    
    def _create_queue(self, n):
        
        return (n * ctypes.py_object)()

    #Inserta en 1 frente
    def enqueue1(self, item):
        
        if self.item_count1 == self.n:
            raise ValueError("no more capacity")
        self.queue[self.item_count1] = item
        self.item_count1 += 1
    
    #Inserta en 2 frente
    def enqueue2(self, item):
        
        self.queue[self.item_count2] = item
        self.item_count2 += 1
    
    # Elimina 1 frente 
    def dequeue1(self):
        
        c = self.queue[0]
        for i in range(1,self.item_count1):
            self.queue[i-1] = self.queue[i]
        self.queue[self.item_count1 - 1] = ctypes.py_object
        self.item_count1 -= 1
        return c
    
    # Elimina 2 frente 
    def dequeue2(self):
       
        c = self.queue[0]
        for i in range(self.item_count2,1):
            self.queue[i-1] = self.queue[i]
        self.queue[self.item_count2 - 1] = ctypes.py_object
        self.item_count2 -= 1
        return c
    
    def size(self):
        return self.item_count1 and self.item_count2
    
    
q = Queue(5)
q.enqueue1(1)
q.enqueue2(4)
q.dequeue2()
q.dequeue1()
q.size()
```
